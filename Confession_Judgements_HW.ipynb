```python
import pandas as pd 
import regex as re 
with open("confession-judgments.txt", "r") as text_obj:
    text = text_obj.read()
```

```python
text
```

```python
### Iterate through cases to capture: case number, date received, total repayment amount, full repayment by date
### Don't forget to group 
```

```python
### case number
case_number_pat = re.compile(r"PM INDEX NO.\s(\d+)", re.M)
request_case_number = case_number_pat.findall(text)
request_case_number
```

```python
### date received
date_received_pat = re.compile(r"RECEIVED NYSCEF:\s(\d{1,2}/\d{1,2}/\d{4})", re.M)
request_date_received = date_received_pat.findall(text)
request_date_received
```

```python
### total repayment amount
total_repayment_pat = re.compile(r"in the sum of[^$]*\$(\d+(?:,\d{3})*\.\d{2})", re.M) 
total_payment_received = total_repayment_pat.findall(text)
total_payment_received
```

```python
### full repayment by date
repayment_date_pat = re.compile(r"repay the assistance in full[^b]*by ([A-Za-z]+ \d{1,2}, \d{4})", re.M)
repayment_date_received = repayment_date_pat.findall(text)
repayment_date_received 
```

```python
df = pd.DataFrame(list(zip(request_case_number, request_date_received, total_payment_received, repayment_date_received )))
df.columns = ["case number", "date received", "total repayment amount", "full repayment by date"] 
df
```
